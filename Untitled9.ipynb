```python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'lifeexpectancy'
table_name = 'lifeexpectancy'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

life_df = pd.read_sql_query('select * from lifeexpectancy',con=engine)

# no need for an open connection, 
# as we're only doing a single query
engine.dispose()
```

```python
life_df.rename(columns={' thinness  1-19 years':'Thinness Ages 10-19'}, inplace=True)
life_df.rename(columns={' thinness 5-9 years':'Thinness Ages 5-9'}, inplace=True)
life_df.rename(columns={' BMI ':'BMI'}, inplace=True)
life_df.rename(columns={'Life expectancy ':'Life expectancy'}, inplace=True)
life_df.rename(columns={'Measels ':'Measels'}, inplace=True)
life_df.rename(columns={'under-five deaths ':'Deaths 5 Years and Younger'}, inplace=True)
life_df.rename(columns={'Diphtheria ':'Diphtheria'}, inplace=True)
life_df.rename(columns={' HIV/AIDS':'HIV/AIDS'}, inplace=True)
life_df.head()
```

```python
print(life_df.isnull().sum())
```

```python
life_df.dropna(subset=['Income composition of resources'], inplace=True)
life_df.dropna(subset=['Schooling'], inplace=True)
life_df.dropna(subset=['Population'], inplace=True)
life_df.dropna(subset=['GDP'], inplace=True)
life_df.dropna(subset=['Total expenditure'], inplace=True)
life_df.dropna(subset=['Hepatitis B'], inplace=True)
life_df.dropna(subset=['Alcohol'], inplace=True)
```

```python
life_df["Life expectancy"].fillna(life_df["Life expectancy"].mean(), inplace=True)
life_df["Adult Mortality"].fillna(life_df["Adult Mortality"].mean(), inplace=True)
life_df["BMI"].fillna(life_df["BMI"].mean(), inplace=True)
life_df["Polio"].fillna(life_df["Polio"].mean(), inplace=True)
life_df["Diphtheria"].fillna(life_df["Diphtheria"].mean(), inplace=True)
life_df["Thinness Ages 10-19"].fillna(life_df["Thinness Ages 10-19"].mean(), inplace=True)
life_df["Thinness Ages 5-9"].fillna(life_df["Thinness Ages 5-9"].mean(), inplace=True)
```

```python
from scipy.stats.mstats import winsorize

winsorized_BMI = winsorize(life_df['BMI'], (0, 0.10))
winsorized_GDP = winsorize(life_df['GDP'], (0, 0.10))
winsorized_alcohol = winsorize(life_df['Alcohol'], (0, 0.10))
winsorized_percentage_expenditure = winsorize(life_df['percentage expenditure'], (0, 0.10))
winsorized_income_composition_of_resources = winsorize(life_df['Income composition of resources'], (0, 0.10))
winsorized_schooling = winsorize(life_df['Schooling'], (0, 0.10))
```

```python
plt.boxplot(winsorized_BMI)
plt.title("Box plot of BMI")
plt.show()

plt.boxplot(winsorized_GDP)
plt.title("Box plot of GDP")
plt.show()

plt.boxplot(winsorized_alcohol)
plt.title("Box plot of alcohol")
plt.show()

plt.boxplot(winsorized_percentage_expenditure)
plt.title("Box plot of percentage expenditure")
plt.show()

plt.boxplot(winsorized_income_composition_of_resources)
plt.title("Box plot of income composition of resources")
plt.show()

plt.boxplot(winsorized_schooling)
plt.title("Box plot of schooling")
plt.show()
```
